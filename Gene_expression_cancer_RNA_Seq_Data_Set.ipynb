*   Samples (instances) are stored row-wise. *Amostras (instâncias) são armazenadas em linha.*
* Variables (attributes) of each sample are RNA-Seq gene expression levels measured by illumina HiSeq platform. *As variáveis (atributos) de cada amostra são os níveis de expressão do gene RNA-Seq medidos pela plataforma Illumina HiSeq.*

> Attribute Information:

* A dummy name (gene_XX) is given to each attribute. Check the original submission ([https://www.synapse.org/#!Synapse:syn4301332), or the platform specs for the complete list of probes name. The attributes are ordered consitently with the original submission. *Um nome fictício (gene_XX) é dado a cada atributo. Verifique o envio original ([https://www.synapse.org/#!Synapse:syn4301332) ou as especificações da plataforma para obter a lista completa de nomes de probes. Os atributos são ordenados de forma consistente com o envio original.*


* Relevant Papers: Weinstein, John N., et al. 'The cancer genome atlas pan-cancer analysis project.' Nature genetics 45.10 (2013): 1113-1120.

# **Estatística Descritiva:**

## *Utilizando Pandas Profiling*

```python
## Pandas Profiling Relatório
from pandas_profiling import ProfileReport

profile = ProfileReport(data, title='Estatística Descritiva RNA-Seq', html={'style':{'full_width':True}})

profile.to_notebook_iframe()

profile.to_file(output_file="Estatística Descritiva RNA-Seq.html")
```

## *Utilizando Dataprep*

```python
# Relatório Automatizado
from dataprep.eda import create_report
create_report(data)
```

```python
# Relatório de Missing
from dataprep.eda import plot_missing
plot_missing(data)
```

```python
# Relatório de Correlação
from dataprep.eda import plot_correlation
plot_correlation(data)
```

```python
# Relatório de Distribuição das Variáveis
from dataprep.eda import plot
plot(data)
```

## *Utilizando Sweetviz*

```python
# Analyzing data via sweetviz
report=sv.analyze(data)

# Generating report
report.show_html('Estatística Descritiva RNA-Seq SWEETVIZ.html')
```
